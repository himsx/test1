```python
#The urllib.request module defines functions and classes which help in opening URLs (mostly HTTP) in a complex world
from urllib.request import urlopen
from bs4 import BeautifulSoup   #scrapy another popular library
import re   #https://oxylabs.io/blog/regex-web-scraping
import pandas as pd
html = urlopen('https://en.wikipedia.org/wiki/Main_Page')  #highly recommend to inspect the page in web browser
bs = BeautifulSoup(html, "html.parser")
print(bs.prettify())
```

```python
# for finding the titles which start with h1 to h6 element, we use the To use the .find_all() method (refer:https://scrapeops.io/python-web-scraping-playbook/python-beautifulsoup-findall/)
titles = bs.find_all(['h1', 'h2'])  #further ref:https://scrapeops.io/python-web-scraping-playbook/python-beautifulsoup-findall/
print('List all the header tags :', *titles, sep='\n\n')
```

# Extract and Store Function Names & Usage
<br>
From the Python Documentation - Random Functions webpage: https://docs.python.org/3/library/random.html


```python
import urllib.request
from bs4 import BeautifulSoup as bs

#load html code from a url
page = urllib.request.urlopen("https://docs.python.org/3/library/random.html") #highly recommend to inspect the page in web browser
soup = bs(page)
print(soup.prettify())
```

```python
#for finding the functions names
names = soup.body.findAll('dt')
print(names)
```

```python
function_names = re.findall('id="random.\w+', str(names))
print(function_names)
function_names = [item[4:] for item in function_names]
print(function_names)   #final functions names what I want as one column in my CSV
```

```python
#find all function descriptions
description = soup.body.findAll('dd')
print(description)
```

```python
function_usage = []

for item in description:
  item = item.text
  function_usage.append(item)

print(function_usage)
```

```python
#for removing the \n from the usage we will update the code as per below
function_usage = []
for item in description:
  item = item.text
  item = item.replace('\n', ' ')
  function_usage.append(item)

print(function_usage)
```

```python
print('list of function names:',function_names[:5])
print('\nfunction description:', function_usage[0])
print('\nnumber of items in function names:', len(function_names))
print('number of items in function description:', len(function_usage))
```

# Store Data inside a DataFrame
<br>
After ensuring the lenghts of both lists match

```python
#create a dataframe
data = pd.DataFrame({'function name': function_names, 'function usage': function_usage})
data.head()
```
